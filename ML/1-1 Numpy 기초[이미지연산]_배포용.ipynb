### 마스킹 연산이 가능합니다.

## Numpy 활용

### 저장 및 불러오기가 가능합니다.

````python
import numpy as np

# 단일 객체 저장 및 불러오기
array = np.arange(0, 10)
np.save('saved.npy', array)

result = np.load('saved.npy')
print(result)
````

````python
import numpy as np

# 복수 객체 저장 및 불러오기
array1 = np.arange(0, 10)
array2 = np.arange(10, 20)
np.savez('saved.npz', array1=array1, array2=array2)

data = np.load('saved.npz')
result1 = data['array1']
result2 = data['array2']
print(result1)
print(result2)
````

### 정렬할 수 있습니다. 

````python
import numpy as np

# Numpy 원소 오름차순 정렬
array = np.array([5, 9, 10, 3, 1])
array.sort()
print(array)

# Numpy 원소 내림차순 정렬
array = np.array([5, 9, 10, 3, 1])
print('정렬전:',array)
array.sort()
print('정렬후:',array[::-1]) # 정렬을 한 후 mirror view 생성

# 축을 기준으로 정렬
array = np.array([[5, 9, 10, 3, 1], [8, 3, 4, 2, 5]])
array.sort(axis=0)
print(array)
````

## 선형대수 연산 
### 벡터의 곱셈

* 내적은 점(dot)으로 표기한다 ${A \cdot B}$
* 두 벡터를 내적하려면 다음과 같은 조건이 만족되어야 한다.
    * 우선 두 벡터의 차원(길이)이 같아야 한다.
    * 앞의 벡터가 행 벡터이고 뒤의 벡터가 열 벡터여야 한다.

````python
x = np.array([1,2,3])
y = np.array([4,5,6])

np.dot(x,y) # 넘파이가 앞의 벡터는 행 백터이고, 뒤의 벡터는 열 벡터라고 가정함
````

````python
import numpy as np
A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[1, 2],
              [3, 4],
              [5, 6]])

dot_product = np.dot(A, B)
print('행렬 내적 결과:\n', dot_product)
````

* 전치 행렬

````python
A = np.array([[1, 2],
              [3, 4]])
transpose_mat = np.transpose(A)
print('A의 전치 행렬:\n', transpose_mat)
````

# OpenCV

* 영상 처리와 컴퓨터 비전을 위한 오픈소스 라이브러리

```python
cv2.imread(file_name, flag) 
```
- file_name: 읽고자 하는 이미지 파일
- flag: 이미지를 읽는 방법 설정
- IMREAD_COLOR: 이미지를 Color로 읽고, 투명한 부분은 무시
- IMREAD_GRAYSCALE: 이미지를 Grayscale로 읽기
- IMREAD_UNCHANGED: 이미지를 Color로 읽고, 투명한 부분도 읽기(Alpha)

### 특정 범위 픽셀 변경

````python
import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('./image/cat.jpg')

start_time = time.time()
for i in range(0, 100):
    for j in range(0, 100):
        image[i, j] = [255, 255, 255]
print("--- %s seconds ---" % (time.time() - start_time))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

start_time = time.time()
image[0:100, 0:100] = [0, 0, 0]
print("--- %s seconds ---" % (time.time() - start_time))

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
````

###  ROI(Region of Interest: 관심 있는 영역) 추출

````python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('./image/cat.jpg')

# Numpy Slicing: ROI 처리 가능
roi = image[200:350, 50:200]
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.show()
# ROI 단위로 이미지 복사하기
image[0:150, 0:150] = roi

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
````

### 알파 블렌딩 합성

````python
import cv2
import numpy as np
import matplotlib.pylab as plt

# 연산에 사용할 이미지 읽기
img1 = cv2.imread('./image/wing_wall.jpg')
img2 = cv2.imread('./image/yate.jpg')

# 합성에 사용할 알파 값
#두 이미지에 동일한 가중치를 주고 합성한다는 뜻입니다.
alpha = 0.5 

# NumPy 배열에 수식을 직접 연산해서 알파 블렌딩 적용
blended = img1 * alpha + img2 * (1-alpha) 
blended = blended.astype(np.uint8) # 소수점 발생을 제거하기 위함
# img4 = cv2.add(img1, img2) # OpenCV 함수

# imgs = {'img1':img1, 'img2':img2, 'img1+img2': img3, 'cv.add(img1, img2)': img4}

plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))

# 이미지 출력
# for i, (k, v) in enumerate(imgs.items()):
#     plt.subplot(2,2, i + 1)
#     plt.imshow(v[:,:,::-1])
#     plt.title(k)
#     plt.xticks([]); plt.yticks([])
plt.axis('off')
plt.show()
````
